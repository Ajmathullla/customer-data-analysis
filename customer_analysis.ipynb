```python
# Mock Customer Data Analysis
This project analyzes customer behavior, order values, delivery performance, and satisfaction using a mock dataset.
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\admin\Desktop\LMRi - Mock Customer Dataset 2.csv")
```

```python
df
```

```python
#Clean Order value
df["Order Value"]=df["Order Value (?)"].str.replace(",","").astype(int)

#convert Order Date to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])
df
```

```python
print(df.describe())
print(df["Delivery Partner"].value_counts())
```

```python
#order value distribution 
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["Order Value"], bins=10, kde=True)
plt.title("Order Value Distribution")
plt.show()
```

```python
#Average Rating per delivery partner 
sns.barplot(x="Delivery Partner", y="Product Rating (1?5)", data=df)
plt.title("Average Rating by Delivery Partner")
plt.show()
```

```python
#complaints by delivery partner 
sns.countplot(x="Delivery Partner", hue="Complaints Raised", data=df)
plt.title("Complaints Distribution by Delivery Partner")
plt.show()
```

```python
#NPS vs rating
sns.scatterplot(x="Product Rating (1?5)", y="NPS", hue="Return Status", data=df)
plt.title("NPS vs Rating (Colored by Return Status)")
plt.show()
```

```python
## 📊 Insights & Conclusion
- Average order value: ₹1,329 | Avg delivery time: 7.2 days | Avg rating: 3.3/5 | Avg NPS: 5.6  
- *Bluedart* → Best performer (high rating 4.7, NPS 8.8, 0% complaints/returns)   
- *Shadowfax* → Worst performer (rating 1.8, NPS 2.8, 100% complaints, 60% returns)   
- *Delhivery* → Affordable but very high complaint rate (~85%).  
- Strong positive correlation (0.96) between rating & NPS → Improving quality boosts loyalty 🚀  
```
